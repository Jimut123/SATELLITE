```python
import seaborn as sns
import matplotlib.pyplot as plt     
array = [[6,0],
         [5,6]]

plt.figure(figsize = (8,8))
ax= plt.subplot()
sns.set(font_scale=3.0) # Adjust to fit
sns.heatmap(array, annot=True, ax = ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted',fontsize = 30);ax.set_ylabel('True',fontsize = 30); 
ax.set_title('Confusion Matrix', fontsize = 30); 
ax.xaxis.set_ticklabels(['Ship', 'No Ship'],fontsize = 30); ax.yaxis.set_ticklabels(['Ship', 'No Ship'],fontsize = 30);
```
